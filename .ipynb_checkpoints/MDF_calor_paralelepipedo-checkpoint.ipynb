```python
#generación de matriz de valores iniciales
def define_matrix(h):
    global xdim
    global ydim
    global zdim
    xi = [x/10 for x in range(int(h*10),xdim*10,int(h*10))]
    yi = [x/10 for x in range(int(h*10),ydim*10,int(h*10))]
    zi = [x/10 for x in range(int(h*10),zdim*10,int(h*10))]
    xdim = len(xi)
    ydim = len(yi)
    zdim = len(zi)
    print("xi:%s \nyi:%s\nzi:%s"%(str(xi),str(yi),str(zi)))
    print("dimensiones xi:%i, yi:%i, zi:%i"%(xdim,ydim,zdim))
        
    A = np.zeros((xdim,ydim,zdim))
    #condiciones de borde:
    for i in range(xdim):
        for k in range(zdim):
            A[i][ydim-1][k] = 10
    print(A)
    b = np.zeros((ydim))
    print("b:%s"%str(b))
```

```python
# segunda derivada parcial con respecto a la primera variable
# parámetros: i,j,k: indices; m: valor del coeficiente que acompaña la derivada
def d2Fdx(i,j,k,m):
    global A
    A[i-1][j][k] = A[i-1][j][k] + m/(h*h)
    A[i][j][k]   = A[i][j][k] - 2*m/(h*h)
    try:
        A[i+1][j][k] = A[i+1][j][k] + m/(h*h)
    except:
        pass
    if(j == 0):
        b[j] = b[j] - m * A[0][0][0]/(h*h)
    if((j+1)== ydim):
        b[j] = b[j] - m * A[i][ydim-1][k]/(h*h)
        
# segunda derivada parcial con respecto a la segunda variable
def d2Fdy(i,j,k,m):
    global A
    A[i][j-1][k] = A[i][j-1][k] + m/(h*h)
    A[i][j][k]   = A[i][j][k] - 2*m/(h*h)
    try:
        A[i][j+1][k] = A[i][j+1][k] + m/(h*h)
    except:
        pass
    if(j == 0):
        b[j] = b[j] - m * A[0][0][0]/(h*h)
    if((j+1)== ydim):
        b[j] = b[j] - m * A[i][ydim-1][k]/(h*h)
        
# segunda derivada parcial con respecto a la tercera variable
def d2Fdz(i,j,k,m):
    global Aijk
    A[i][j][k-1] = A[i][j][k-1] + m/(h*h)
    A[i][j][k]   = A[i][j][k] - 2*m/(h*h)
    try:
        A[i][j][k+1] = A[i][j][k+1] + m/(h*h)
    except:
        pass
    if(j == 0):
        b[j] = b[j] - m * A[0][0][0]/(h*h)
    if((j+1)== ydim):
        b[j] = b[j] - m * A[i][ydim-1][k]/(h*h)
```
